```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

for col in num_cols:
    diabetes = remove_outliers(diabetes, col)
```

```python
# ===============================
# DATA PREPROCESSING: Diabetes
# ===============================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# 1️⃣ Load dataset
diabetes_df = pd.read_csv("Diabetes .csv")

# 2️⃣ Handle missing values
# Replace 0 in BMI with NaN (if any), then fill numeric missing values with median
diabetes_df['BMI'] = diabetes_df['BMI'].replace(0, np.nan)
numeric_cols = diabetes_df.select_dtypes(include=np.number).columns
diabetes_df[numeric_cols] = diabetes_df[numeric_cols].fillna(diabetes_df[numeric_cols].median())

# 3️⃣ Handle categorical data
# Encode all object type columns (Gender)
encoder = LabelEncoder()
for col in diabetes_df.select_dtypes(include='object').columns:
    diabetes_df[col] = encoder.fit_transform(diabetes_df[col])

# 4️⃣ Handle outliers using IQR method
Q1 = diabetes_df[numeric_cols].quantile(0.25)
Q3 = diabetes_df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filter rows that are NOT outliers
outlier_condition = ((diabetes_df[numeric_cols] < (Q1 - 1.5 * IQR)) |
                     (diabetes_df[numeric_cols] > (Q3 + 1.5 * IQR)))
diabetes_df = diabetes_df[~outlier_condition.any(axis=1)]

# 5️⃣ Separate features and target
X = diabetes_df.drop(['CLASS', 'ID', 'No_Pation'], axis=1)
y = diabetes_df['CLASS']

# 6️⃣ Min-Max Normalization
X_minmax = pd.DataFrame(MinMaxScaler().fit_transform(X), columns=X.columns)
diabetes_minmax = X_minmax.copy()
diabetes_minmax['CLASS'] = y.values

# 7️⃣ Standard Scaling
X_standard = pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns)
diabetes_standard = X_standard.copy()
diabetes_standard['CLASS'] = y.values

# 8️⃣ Display results
print("--- Diabetes Min-Max Normalized Data ---")
display(diabetes_minmax.head())

print("\n--- Diabetes Standard Scaled Data ---")
display(diabetes_standard.head())

# 9️⃣ Display shapes after cleaning
print("\nShape after cleaning:", diabetes_df.shape)
```

```python

# Load dataset
adult_df = pd.read_csv("/content/adult.csv")

# Handle missing values
adult_df.replace(" ?", np.nan, inplace=True)
for col in adult_df.select_dtypes(include="object").columns:
    adult_df[col].fillna(adult_df[col].mode()[0], inplace=True)

# Handle categorical data
encoder = LabelEncoder()
for col in adult_df.select_dtypes(include="object").columns:
    adult_df[col] = encoder.fit_transform(adult_df[col])

# Handle outliers
Q1 = adult_df.quantile(0.25)
Q3 = adult_df.quantile(0.75)
IQR = Q3 - Q1
adult_df = adult_df[~((adult_df < (Q1 - 1.5 * IQR)) |
                        (adult_df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Min-Max Normalization
adult_minmax = pd.DataFrame(
    MinMaxScaler().fit_transform(adult_df),
    columns=adult_df.columns
)

# Standard Scaling
adult_standard = pd.DataFrame(
    StandardScaler().fit_transform(adult_df),
    columns=adult_df.columns
)

print("--- Adult Min-Max Normalized Data (Head) ---")
print(adult_minmax.head())
print("\n--- Adult Standard Scaled Data (Head) ---")
print(adult_standard.head())
```
