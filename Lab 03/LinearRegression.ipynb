```python
#canada_per_captia_income
import pandas as pd
from sklearn.linear_model import LinearRegression

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("canada_per_capita_income.csv")

X = df[['year']]
y = df['per capita income (US$)']

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[2020]])

print("Predicted value for year 2020:",prediction)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income - Linear Regression")
plt.show()
```

```python

#salary.csv
import pandas as pd
from sklearn.linear_model import LinearRegression
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("salary.csv")

print(df.isnull().sum())
mean_experience = df['YearsExperience'].mean()

df['YearsExperience'].fillna(mean_experience, inplace=True)

X = df[['YearsExperience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)

predicted_salary = model.predict([[12]])

print(predicted_salary)
```
