```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('canada_per_capita_income.csv')

# Display first few rows
print(df.head())

# Check for missing values
print(df.isnull().sum())

# (Optional) Rename column if it has spaces
df.rename(columns={'per capita income (US$)': 'income'}, inplace=True)

# Define input (X) and output (y)
X = df[['year']]      # Independent variable must be 2D
y = df['income']      # Dependent variable

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict income for year 2020
predicted_income = model.predict([[2020]])

print("Predicted per capita income for Canada in 2020: $", predicted_income[0])
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('salary.csv')

# Display first few rows
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Data Cleaning
# Remove rows where YearsExperience or Salary is missing
df = df.dropna()

# Ensure numeric format
df['YearsExperience'] = pd.to_numeric(df['YearsExperience'])
df['Salary'] = pd.to_numeric(df['Salary'])

# Define independent and dependent variables
X = df[['YearsExperience']]   # Must be 2D
y = df['Salary']

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict salary for 12 years experience
predicted_salary = model.predict([[12]])

print("Predicted Salary for 12 years experience: $", round(predicted_salary[0], 2))
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('hiring.csv')

print("Original Data:")
print(df)

# -------------------------
# DATA PREPROCESSING
# -------------------------

# Convert word numbers in experience column to numeric
# Mapping dictionary
word_to_num = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11,
    'twelve': 12
}

# Fill missing experience with 'zero'
df['experience'] = df['experience'].fillna('zero')

# Convert words to numbers
df['experience'] = df['experience'].apply(
    lambda x: word_to_num[x] if x in word_to_num else int(x)
)

# Fill missing test_score with median
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].median()
)

print("\nCleaned Data:")
print(df)

# -------------------------
# MODEL TRAINING
# -------------------------

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

print("\nModel Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

# -------------------------
# PREDICTIONS
# -------------------------

# 1) 2 yr experience, 9 test score, 6 interview score
salary1 = model.predict([[2, 9, 6]])

# 2) 12 yr experience, 10 test score, 10 interview score
salary2 = model.predict([[12, 10, 10]])

print("\nPredicted salary for (2 yr exp, 9 test, 6 interview): $", round(salary1[0], 2))
print("Predicted salary for (12 yr exp, 10 test, 10 interview): $", round(salary2[0], 2))
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('1000_Companies.csv')

print("Original Data:")
print(df.head())

# -----------------------------
# DATA PREPROCESSING
# -----------------------------

# One Hot Encoding for 'State'
df = pd.get_dummies(df, columns=['State'], drop_first=True)

print("\nData After One-Hot Encoding:")
print(df.head())

# Define independent and dependent variables
X = df.drop('Profit', axis=1)
y = df['Profit']

# Create model
model = LinearRegression()
model.fit(X, y)

print("\nModel Coefficients:")
print(model.coef_)

print("\nModel Intercept:")
print(model.intercept_)

# -----------------------------
# PREDICTION
# -----------------------------

# Input:
# 91694.48 R&D Spend
# 515841.3 Administration
# 11931.24 Marketing Spend
# Florida State

# IMPORTANT:
# After drop_first=True, the states become:
# State_Florida
# State_New York
# (California becomes base case)

# For Florida:
# State_Florida = 1
# State_New York = 0

prediction = model.predict([[91694.48, 515841.3, 11931.24, 1, 0]])

print("\nPredicted Profit: $", round(prediction[0], 2))
```
